```python
# Check given number is prime or not
def is_prime(n):
  if (n < 0):
    print(f"{n} is not a prime number")
    return
  for i in range(2, int(n**0.5)+1):
    if n%i==0:
      print(f"{n} is not a prime number")
      return
  print(f"{n} is a prime number")

is_prime(9)  # True
is_prime(10)  # False


```

```python
# Check all the numbers in the range is prime or not
def check_prime(n):
  for num in range(2, n+1):
    for i in (2, int(num**0.5)+1):
      if (num % i == 0):
        print(f"{num} is not a prime")
        break
    else:
      print(f"{num} is a prime")

check_prime(9)
```

```python
# Factorial
def check_factorial(n):
  fact = 1
  for i in range(2, n+1):
    fact *= i
  print(fact)

check_factorial(7)

```

```python
# Strings

# Reverse a string
s = "Sathish"
print(f"String Reverse: {s[::-1]}")

# Check palindrome

def check_palindrome(s1):
  if (s1.lower() == s1[::-1].lower()):
    print(f"Given string {s} is Palindrome")
  else:
    print(f"Given string {s} is not Palindrome")

check_palindrome("Madam")

# Count characters in the string
from collections import Counter

print(Counter("Sathish Kumar"))

def check_count(s):
  dict = {}
  for i in s:
    dict[i] = dict.get(i, 0) +1
  return dict

print(check_count("Sathish"))


# Anagram

def check_anagram(s1, s2):
  if(sorted(s1.lower()) == sorted(s2.lower())):
    print(f"Given strings {s1} and {s2} are Anagram")
  else:
    print(f"Given strings {s1} and {s2} are not Anagram")

check_anagram("silent", "listen")

# 2nd non repeating char in a string
s3 = "Programming"
count = Counter(s3)
print(count)
print([ch for ch in s3 if count[ch] ==1][1])



```
